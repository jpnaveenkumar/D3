```python
import pandas as pd
```

```python
test_match_results_19 = pd.read_csv ('Men Test Team Match Results - 19th Century.csv')
test_match_results_20 = pd.read_csv ('Men Test Team Match Results - 20th Century.csv')
test_match_results_21 = pd.read_csv ('Men Test Team Match Results - 21st Century.csv')
```

```python
test_match_results = pd.concat([test_match_results_19, test_match_results_20, test_match_results_21])
```

```python
test_match_results.head()
```

```python
test_player_innings_19 = pd.read_csv ('Men Test Player Innings Stats - 19th Century.csv')
test_player_innings_20 = pd.read_csv ('Men Test Player Innings Stats - 20th Century.csv')
test_player_innings_21 = pd.read_csv ('Men Test Player Innings Stats - 21st Century.csv')
```

```python
test_player_innings = pd.concat([test_player_innings_19, test_player_innings_20, test_player_innings_21])
```

```python
test_player_innings.head()
```

```python
test_bowling_stats = pd.read_csv ('Men Test Team Bowling Stats.csv')
test_batting_stats = pd.read_csv ('Men Test Team Batting Stats.csv')
```

```python
test_bowling_stats.head()
```

```python
test_batting_stats.head()
```

```python
odi_match_results_20 = pd.read_csv ('Men ODI Team Match Results - 20th Century.csv')
odi_match_results_21 = pd.read_csv ('Men ODI Team Match Results - 21st Century.csv')
```

```python
odi_match_results = pd.concat([odi_match_results_20, odi_match_results_21])
```

```python
odi_match_results.head()
```

```python
odi_player_innings_20 = pd.read_csv ('Men ODI Player Innings Stats - 20th Century.csv')
odi_player_innings_21 = pd.read_csv ('Men ODI Player Innings Stats - 21st Century.csv')
```

```python
odi_player_innings = pd.concat([odi_player_innings_20, odi_player_innings_21])
```

```python
odi_player_innings.head()
```

```python
odi_bowling_stats = pd.read_csv ('Men ODI Team Bowling Stats.csv')
odi_batting_stats = pd.read_csv ('Men ODI Team Batting Stats.csv')
```

```python
odi_bowling_stats.head()
```

```python
odi_batting_stats.head()
```

```python
test_match_results.to_csv('Men Test Team Match Results.csv')
test_player_innings.to_csv('Men Test Player Innings Stats.csv')
odi_match_results.to_csv('Men ODI Team Match Results.csv')
odi_player_innings.to_csv('Men ODI Player Innings Stats.csv')
```
